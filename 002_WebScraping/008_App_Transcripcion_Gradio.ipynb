# Aplicación de Transcripción de Audio con OpenAI Whisper

Este cuaderno utiliza la librería Gradio para crear una interfaz gráfica que permite a l@s usuari@s cargar archivos de audio y transcribir su contenido utilizando el modelo "Whisper" de OpenAI.

## Importación de Librerías

A continuación, importamos las librerías necesarias para nuestro proyecto.

```python
import torch
from transformers import pipeline
import gradio as gr
```

## Función para Transcribir Audio

Definimos una función que se encargará de transcribir el audio utilizando el modelo Whisper.

```python
# Función para transcribir audio utilizando el modelo OpenAI Whisper
def transcript_audio(audio_file):
    # Inicializa el pipeline de reconocimiento de voz
    pipe = pipeline(
        "automatic-speech-recognition",
        model="openai/whisper-small",
        chunk_length_s=30,
    )

    # Transcribe el archivo de audio y devuelve el resultado
    result = pipe(audio_file, batch_size=8)["text"]
    return result
```

## Configuración de la Interfaz Gradio

Creamos la interfaz de Gradio, que incluye un componente de carga de audio y un cuadro de texto para mostrar la transcripción resultante.

```python
# Entrada de audio
audio_input = gr.Audio(sources="upload", type="filepath")  # Permitir carga de archivo de audio
# Salida de texto
output_text = gr.Textbox()  # Cuadro de texto para mostrar la transcripción

# Crear la interfaz de Gradio con la función, entradas y salidas
iface = gr.Interface(fn=transcript_audio,
                     inputs=audio_input, outputs=output_text,
                     title="Aplicación de Transcripción de Audio",
                     description="Carga el archivo de audio para transcribirlo")
```

## Lanzar la Aplicación Gradio

Finalmente, lanzamos la aplicación Gradio para que los usuarios puedan interactuar con ella.

```python
# Lanza la aplicación Gradio
iface.launch(inline=True)
```

## Conclusión

Creamos una aplicación simple que permite transcribir audio a texto utilizando el modelo Whisper de OpenAI y la interfaz de usuario de Gradio.
